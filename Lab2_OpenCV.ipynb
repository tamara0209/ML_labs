```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt
```

```python
# 1. Resize 

image = cv2.imread('/Users/ts/Desktop/paris.jpg',cv2.IMREAD_UNCHANGED)
print('Original Dimensions : ',image.shape)
 
width = 200
height = 300
dim = (width, height)

resized = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)
 
print('Resized Dimensions : ',resized.shape)
 
cv2.imshow("Resized image", resized)
cv2.imwrite("resized.jpg", resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# 2. Rotate

image = cv2.imread('/Users/ts/Desktop/paris.jpg')
(h, w) = image.shape[:2]
center = (w / 2, h / 2)
 
angle90 = 90
angle180 = 180
angle270 = 270
 
scale = 1.0
 
M = cv2.getRotationMatrix2D(center, angle90, scale)
rotated90 = cv2.warpAffine(image, M, (h, w))
 
M = cv2.getRotationMatrix2D(center, angle180, scale)
rotated180 = cv2.warpAffine(image, M, (w, h))

M = cv2.getRotationMatrix2D(center, angle270, scale)
rotated270 = cv2.warpAffine(image, M, (h, w))
 
 
cv2.imshow('Original Image',image)
cv2.imshow('Image rotated by 90 degrees',rotated90)
cv2.imshow('Image rotated by 180 degrees',rotated180)
cv2.imshow('Image rotated by 270 degrees',rotated270)
cv2.waitKey(0) 
cv2.destroyAllWindows() 
```

```python
# 3

import cv2
import numpy as np
from matplotlib import pyplot as plt

originalImage = cv2.imread('/Users/ts/Desktop/paris.jpg',cv2.IMREAD_UNCHANGED)
image = cv2.cvtColor(originalImage, cv2.COLOR_BGR2GRAY)

ret,thresh1 = cv2.threshold(image,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(image,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(image,127,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(image,127,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(image,127,255,cv2.THRESH_TOZERO_INV)

titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [image, thresh1, thresh2, thresh3, thresh4, thresh5]

for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```

```python
img = cv2.imread("/Users/ts/Desktop/paris.jpg")
rows, cols, ch = img.shape
print("Height: ", rows)
print("Width: ", cols)


scaled_img = cv2.resize(img, (60, 60))

matrix_t = np.float32([[1, 0, -100], [0, 1, -30]])
translated_img = cv2.warpAffine(img, matrix_t, (cols, rows))


matrix_r = cv2.getRotationMatrix2D((cols/2, rows/2), 90, 0.5)
rotated_img = cv2.warpAffine(img, matrix_r, (cols, rows))


cv2.imshow("Original image", img)
cv2.imshow("Scaled image", scaled_img)
cv2.imshow("Translated image", translated_img)
cv2.imshow("Rotated image", rotated_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
